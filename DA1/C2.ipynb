```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as rd
import sys
import seaborn as sns
import squarify as sq
import unicodedata
```

```python
data=pd.read_csv(r"C:\Users\sujit\Downloads\newautos.csv", encoding_errors= 'replace')
```

```python
datac=data.copy()
```

```python
#2.What is the Average price of vehicle by fuel type and gearbox type.Give a plot
```

```python

average_price = datac.groupby(['fuelType', 'gearbox'])['price'].mean().unstack()
average_price.plot(kind='bar', stacked=True,figsize=(5,6))
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.title('Average Price of Vehicles by Fuel Type and Gearbox Type')
plt.xticks(rotation=45)
plt.legend(title='Gearbox Type')
plt.show()
```
