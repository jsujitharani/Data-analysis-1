```python
# 1) Can you tell me No of Vehicles by Brand Available on ebay for sale with the help of visualization
```

```python
datac
```

```python
g=datac.groupby("brand")
```

```python
a=g[["vehicleType"]].count()
```

```python
da=pd.DataFrame(a.index)
```

```python
da["count"]=a.values
da
```

```python
plt.figure(figsize=(10, 6))
a=da["brand"]
b=da["count"]

plt.bar(a, b, color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('brand')
plt.ylabel('no.of vehicles')
plt.title('No of Vehicles by Brand Available on ebay for sale')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
