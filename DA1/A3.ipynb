```python
data=pd.read_csv(r"C:\Users\sujit\Downloads\newautos.csv", encoding_errors= 'replace')
```

```python
#3) Create a plot based on the Variation of the price range by the vehicle type


plt.figure(figsize=(10, 6))
a=datac["vehicleType"]
b=datac["price"]
plt.bar(a, b, color='skyblue', alpha=0.7)
plt.xticks(rotation=45)
plt.xlabel('Vehicle Type')
plt.ylabel('Price Range')
plt.title('Variation of Price Range by Vehicle Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
