```python
#4) Find out Total count of vehicles by type available on ebay for sale.As well as create a visualization for the client
```

```python
g=datac.groupby("vehicleType")
```

```python
datac["vehicleType"].unique()
```

```python
a=g[["vehicleType"]].count()
```

```python
a
```

```python
da=pd.DataFrame(a.index)
```

```python

da["count"]=a.values
```

```python
da
```

```python

plt.figure(figsize=(10, 6))
a=da["vehicleType"]
b=da["count"]

plt.bar(a, b, color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Vehicle Type')
plt.ylabel('count of vehicles available for sale')
plt.title('Total count of vehicles by type available on ebay for sale')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
